# Column Analysis

## Column 'Unnamed: 0'   
Doesn't add any information and I decided to drop it

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

## Column 'timestamp'
This column doesn't add information but we need it to buid our target: 'Failure'


```python
# Converting 'timestamp' to datatime type
df['timestamp'] = pd.to_datetime(df['timestamp'], errors="coerce")
```

```python
# Creating a data frame with the failure periods of time
failures = pd.DataFrame({
    'start': pd.to_datetime([
        '2020-04-18 00:00',
        '2020-05-29 23:30',
        '2020-06-05 10:00',
        '2020-07-15 14:30'
    ]),
    'end': pd.to_datetime([
        '2020-04-18 23:59',
        '2020-05-30 06:00',
        '2020-06-07 14:30',
        '2020-07-15 19:00'
    ])
})
```

```python
# Creating subcolumns with 'month', 'day', 'hour', 'minute', 'second'
df['year'] = df['timestamp'].dt.year
df['day'] = df['timestamp'].dt.day
df['month'] = df['timestamp'].dt.month
df['day'] = df['timestamp'].dt.day
df['hour'] = df['timestamp'].dt.hour
df['minute'] = df['timestamp'].dt.minute
df['second'] = df['timestamp'].dt.second
```

```python
df['year'].value_counts()
```

```python
# Droping 'year' because there is only one value
df.drop('year', axis=1, inplace=True)
```

```python
df['month'].value_counts()
```

## Column 'failure'

```python
# Create failure column (default = 0)
df['failure'] = 0

# Mark rows that fall within any failure window as 1
for _, row in failures.iterrows():
    mask = (df['timestamp'] >= row['start']) & (df['timestamp'] <= row['end'])
    df.loc[mask, 'failure'] = 1
```

```python
print(df['failure'].value_counts())
print(f"Failure rate: {df['failure'].mean()*100:.2f}%")
```

## Column 'TP2'  
Measures the pressure on the compressor in bar. Numerical continuous


```python
df['TP2'].describe()
```

## Column 'TP3'  
Mesures the pressure generated at the pneumatic panel in bar. Numerical continuous

```python
df['TP3'].describe()
```

## Column 'H1'   
Measures the pressure generated due to pressure drop when the discharge of the cyclonic separator filter occurs.

```python
df['H1'].describe()
```

## Column 'DV_pressure'  
Measures in bar the pressure drop generated when the towers discharge air dryers; a zero reading indicates that the compressor is operating under load. Numerical continuous

```python
df['DV_pressure'].describe()
```

## Column 'Reservoirs'  
Measures in bar the downstream pressure of the reservoirs, which should be close to the pneumatic panel pressure (TP3). Numerical continuous

```python
df['Reservoirs'].describe()
```

## Column 'Oil_temperature'  
measures in ºC of the oil temperature on the compressor. Numerical continous

```python
df['Oil_temperature'].describe()
```

## Column 'Motor_current'  
measures in A the current of one phase of the three-phase motor; it presents values close to 0A - when it turns off, 4A - when working offloaded, 7A - when working under load, and 9A - when it starts working. Numerical continous

```python
df['Motor_current'].describe()
```

## Column 'COMP'  
electrical signal of the air intake valve on the compressor; it is active when there is no air intake, indicating that the compressor is either turned off or operating in an offloaded state.

```python
df['COMP'].value_counts()
```

## Column 'DV_eletric'  
electrical signal that controls the compressor outlet valve; it is active when the compressor is functioning under load and inactive when the compressor is either off or operating in an offloaded state.

```python
df['DV_eletric'].value_counts()
```

## Column 'Towers'  
electrical signal that defines the tower responsible for drying the air and the tower responsible for draining the humidity removed from the air; when not active, it indicates that tower one is functioning; when active, it indicates that tower two is in operation.

```python
df['Towers'].value_counts()
```

## Column 'MPG'  
electrical signal responsible for starting the compressor under load by activating the intake valve when the pressure in the air production unit (APU) falls below 8.2 bar; it activates the COMP  sensor, which assumes the same behaviour as the MPG sensor.

```python
df['MPG'].value_counts()
```

## Column 'LPS'  
electrical signal that detects and activates when the pressure drops below 7 bars. We need to predict a fealure before this signal works

```python
df['LPS'].value_counts()
```

## Column 'Pressure_switch'  
electrical signal that detects the discharge in the air-drying towers.

```python
df['Pressure_switch'].value_counts()
```

## Column 'Oil_level'  
electrical signal that detects the oil level on the compressor; it is active when the oil is below the expected values.

```python
df['Oil_level'].value_counts()
```

## Column 'Caudal_impulses'  
electrical signal that counts the pulse outputs generated by the absolute amount of air flowing from the APU to the reservoirs.

```python
df['Caudal_impulses'].value_counts()
```

# Univariate Anlaysis

## Numerical Columns

```python
numeric_cols = ['TP2', 'TP3', 'H1', 'DV_pressure', 'Reservoirs', 
                'Oil_temperature', 'Motor_current']

fig, axes = plt.subplots(len(numeric_cols), 2, figsize=(14, len(numeric_cols)*4))

for i, col in enumerate(numeric_cols):
    # Histogram
    sns.histplot(df[col], ax=axes[i, 0], kde=True, color='steelblue')
    axes[i, 0].set_title(f'{col} - Distribution')
    
    # Boxplot
    sns.boxplot(x=df[col], ax=axes[i, 1], color='steelblue')
    axes[i, 1].set_title(f'{col} - Boxplot')

plt.tight_layout()
plt.show()
```

## Categorical Columns

```python
binary_cols = ['COMP', 'DV_eletric', 'Towers', 'MPG', 
               'LPS', 'Pressure_switch', 'Oil_level', 'Caudal_impulses']

fig, axes = plt.subplots(2, 4, figsize=(16, 8))
axes = axes.flatten()

for i, col in enumerate(binary_cols):
    sns.countplot(x=df[col], ax=axes[i], color='steelblue')
    axes[i].set_title(f'{col} - Value Counts')

plt.tight_layout()
plt.show()
```

# Bivariate Anlaysis

## Failure vs Normal

```python
fig, axes = plt.subplots(len(numeric_cols), 1, figsize=(14, len(numeric_cols)*4))

for i, col in enumerate(numeric_cols):
    sns.boxplot(
        x='failure', 
        y=col, 
        data=df, 
        ax=axes[i],
        hue='failure',          # fix for FutureWarning
        palette={0: 'steelblue', 1: 'red'},
        legend=False            # fix for FutureWarning
    )
    axes[i].set_title(f'{col} — Normal vs Failure')
    axes[i].set_xticks([0, 1])                              # fix for UserWarning
    axes[i].set_xticklabels(['Normal (0)', 'Failure (1)'])  # fix for UserWarning

plt.tight_layout()
plt.show()
```

## One week before failure

```python
# Define one week before each failure start
pre_failure_windows = pd.DataFrame({
    'start': pd.to_datetime([
        '2020-04-11 00:00',   # 1 week before failure 1
        '2020-05-22 23:30',   # 1 week before failure 2
        '2020-05-29 10:00',   # 1 week before failure 3
        '2020-07-08 14:30'    # 1 week before failure 4
    ]),
    'end': pd.to_datetime([
        '2020-04-18 00:00',   # failure 1 starts
        '2020-05-29 23:30',   # failure 2 starts
        '2020-06-05 10:00',   # failure 3 starts
        '2020-07-15 14:30'    # failure 4 starts
    ])
})

# Create a new column with 3 categories
df['period'] = 'normal'

for _, row in pre_failure_windows.iterrows():
    mask = (df['timestamp'] >= row['start']) & (df['timestamp'] < row['end'])
    df.loc[mask, 'period'] = 'pre_failure'

df.loc[df['failure'] == 1, 'period'] = 'failure'

print(df['period'].value_counts())
```

```python
# The 3 perios together
fig, axes = plt.subplots(len(numeric_cols), 1, figsize=(14, len(numeric_cols)*4))

for i, col in enumerate(numeric_cols):
    sns.boxplot(
        x='period',
        y=col,
        data=df,
        ax=axes[i],
        hue='period',                                          # add this
        order=['normal', 'pre_failure', 'failure'],
        hue_order=['normal', 'pre_failure', 'failure'],        # add this
        palette={'normal': 'steelblue', 'pre_failure': 'orange', 'failure': 'red'},
        legend=False                                           # add this
    )
    axes[i].set_title(f'{col} — Normal vs Pre-Failure vs Failure')

plt.tight_layout()
plt.show()
```
